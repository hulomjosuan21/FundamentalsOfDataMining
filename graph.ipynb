```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
n_data = data.dropna()
```

```python
groups = n_data[['Sex','HeartDisease']]
group_data = groups.groupby('Sex')['HeartDisease'].mean()
group_data = group_data.sort_values()

fig, ax = plt.subplots(figsize=(16,10), dpi=80)
sns.barplot(x=group_data.index, y=group_data.values)
ax.set_title('ChestPainType Vs Heart Disease', size=20)
ax.set_xlabel('Sex',size=20)
ax.set_ylabel('Avg no. of heart disease', size=20)
ax.set_xticks(ticks=range(len(group_data)), labels=group_data.index, size=20)
```
